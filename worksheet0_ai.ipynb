```python
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

low = []
moderate = []
high = []

for s, e, sl in time_data:
    if s < 3:
        low.append(s)
    elif 3 <= s <= 5:
        moderate.append(s)
    else:
        high.append(s)

print(low)
print(moderate)
print(high)
```

```python
n = int(input("Enter number of days: "))
low = moderate = high = 0

for i in range(n):
    s = float(input(f"Enter study hours for day {i+1}: "))

    if s < 3:
        low += 1
    elif 3 <= s <= 5:
        moderate += 1
    else:
        high += 1

print("\nLow count:", low)
print("Moderate count:", moderate)
print("High count:", high)
```

```python


n = int(input("Enter number of days: "))
study_minutes = []

for i in range(n):
    s = float(input(f"Enter study hours for day {i+1}: "))
    study_minutes.append(s * 60)

print("\nStudy minutes:", study_minutes)
```

```python
n = int(input("Enter number of days: "))

study_hours = []
entertainment_hours = []
sleep_hours = []

for i in range(n):
    print(f"\nDay {i+1}")
    s = float(input("  Study hours: "))
    e = float(input("  Entertainment hours: "))
    sl = float(input("  Sleep hours: "))

    study_hours.append(s)
    entertainment_hours.append(e)
    sleep_hours.append(sl)

print("\nAverage study:", sum(study_hours)/len(study_hours))
print("Average entertainment:", sum(entertainment_hours)/len(entertainment_hours))
print("Average sleep:", sum(sleep_hours)/len(sleep_hours))
```

```python
import matplotlib.pyplot as plt

n = int(input("Enter number of days: "))

study = []
sleep = []

for i in range(n):
    print(f"\nDay {i+1}")
    s = float(input("  Study hours: "))
    sl = float(input("  Sleep hours: "))

    study.append(s)
    sleep.append(sl)

plt.scatter(study, sleep)
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study vs Sleep Pattern")
plt.show()
```
