```python
import matplotlib.pyplot as plt
import random

x = [random.randint(0, 50) for _ in range(10)]
y = [random.randint(0, 50) for _ in range(10)]
colors = ["red", "green", "blue", "yellow", "pink", "orange", "purple", "brown", "cyan", "magenta"]

plt.scatter(x, y, c=colors)
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.show()
```

```python
import matplotlib.pyplot as plt

years = ['2017', '2018', '2019', '2020', '2021', '2022']
population = [150, 180, 210, 250, 290, 330]

plt.pie(population, labels=years, autopct='%1.1f%%')
plt.title("Animal Population Over Years")
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))

for i, ax in enumerate(axes.flatten()):
    ax.pie([population[i]], labels=[years[i]], autopct='%1.1f%%')
    ax.set_title(f"Year {years[i]}")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = [ "Red", "Silver", "Blue", "White", "Green"]
counts = [123, 230, 78, 193, 12]

plt.bar(colors, counts)
plt.xlabel("Car Colors")
plt.xlabel("Car colors seen")
plt.show()

plt.bar(colors, counts, width= 0.3)
plt.show()

plt.barh(colors, counts)
plt.show()

bar_colors= ["red","silver", "blue", "white", "green"]
plt.bar(colors, counts, color=bar_colors)
plt.show()


plt.pie(counts, labels=colors, autopct='%1.1f%%')
plt.title("Car Colors Distribution")
plt.show()



# A. Exploded Pie Chart
explode = [0.3, 0, 0, 0, 0]
plt.pie(counts, labels=colors, autopct='%1.1f%%', explode=explode)
plt.show()

# B. Explode + Shadow
plt.pie(counts, labels=colors, autopct='%1.1f%%', explode=explode, shadow=True)
plt.show()

# C. Adding a Legend
plt.pie(counts, labels=colors, autopct='%1.1f%%')
plt.legend(title="Car Colours:")
plt.show()

# D. Custom Colors
custom_colors = ["magenta", "green", "cyan", "red", "blue"]
plt.pie(counts, labels=colors, autopct='%1.1f%%', colors=custom_colors)
plt.show()

```

```python

# Workshop2

























from math import pi, sin

# Example 1: Convert degrees to radians and compute sine
# Approach 1
def deg_sin():
    x = float(input("Please input a number: "))
    answer = sin(pi * x / 180)
    print("Result =", answer)
deg_sin()

# Approach 2
def deg_sin(x):
    answer = sin(pi * x / 180)
    print("Result =", answer)
inp = float(input("Please input a number: "))
deg_sin(inp)

# Approach 3
def deg_sin(x):
    return sin(pi * x / 180)
inp = float(input("Please input a number: "))
out = deg_sin(inp)
print("Result =", out)

# Exercise 1: Unique elements in a list
def unique_list(lst):
    return list(set(lst))
print(unique_list([1,2,3,3,3,3,4,5]))

# Exercise 2: Check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(29))

# Exercise 3: Twin primes less than 1000
def twin_primes():
    primes = [i for i in range(2, 1000) if is_prime(i)]
    twins = [(primes[i], primes[i+1]) for i in range(len(primes)-1) if primes[i+1] - primes[i] == 2]
    print(twins)
twin_primes()

# Exercise 4: Sum of divisors
def sum_of_divisors(n):
    return sum([i for i in range(1, n) if n % i == 0])
print(sum_of_divisors(6))

# Exercise 5: Find next perfect number
def next_perfect_number(start=7):
    n = start
    while True:
        if sum_of_divisors(n) == n:
            return n
        n += 1
print(next_perfect_number())

# Exercise 6: Mean and Median
def mean_median(marks):
    marks.sort()
    mean = sum(marks) / len(marks)
    mid = len(marks) // 2
    median = (marks[mid] if len(marks) % 2 != 0 else (marks[mid - 1] + marks[mid]) / 2)
    return mean, median
print(mean_median([20, 30, 35, 55, 70, 80, 90]))

# Exercise 7: Decimal to Binary
def decimal_to_binary(n):
    return bin(n)[2:]
print(decimal_to_binary(25))

# Exercise 8: Product of digits
def prod_digits(n):
    prod = 1
    for digit in str(n):
        prod *= int(digit)
    return prod
print(prod_digits(1234))

# Exercise 9: Multiply arbitrary arguments
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result
print(multiply(1,2,3,4,5))
print(multiply(12,13,14))

# Set Operations
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}
U = set(range(1, 51))

print(len(set1))  # Length of set1
print(sum(set2))  # Sum of elements in set2
print(set1 | set2)  # Union
print(set1 & set2)  # Intersection
print(set1 - set2)  # Difference set1 - set2
print(set2 - set1)  # Difference set2 - set1
print(set3.issubset(set1))  # Check if set3 is subset
print(U - set1)  # Complement of set1 in U
print(15 in set1)  # Check if 15 is in set1
set1.add(60)  # Add 60 to set1
set2.discard(60)  # Remove 60 from set2
set3.clear()  # Remove all elements from set3
print(max(set2), min(set2))  # Max and Min of set2
```
