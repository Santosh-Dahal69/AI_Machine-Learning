```python
import numpy as np

size = int(input("Enter size of list: "))

new_list = []
for i in range(size):
    val = float(input(f"Enter element {i+1}: "))
    new_list.append(val)

np_array = np.array(new_list)

print("\nList entered:", new_list)
print("Converted NumPy array:", np_array)
```

```python
import numpy as np

size = int(input("Enter size of array (e.g., 10): "))
pos = int(input("Enter position to replace (e.g., 5): "))

arr = np.zeros(size)
arr[pos - 1] = 1

print("\nModified array:")
print(arr)
```

```python
import numpy as np

size = int(input("Enter array size: "))

arr = []
for i in range(size):
    val = float(input(f"Enter element {i+1}: "))
    arr.append(val)

arr = np.array(arr)
reversed_arr = arr[::-1]

print("\nOriginal array:", arr)
print("Reversed array:", reversed_arr)
```

```python
import numpy as np

rows = int(input("Rows: "))
cols = int(input("Columns: "))

arr = np.ones((rows, cols))
arr[1:-1, 1:-1] = 0

print("\nBorder matrix:")
print(arr)
```

```python
import numpy as np

size = int(input("Enter size for checkerboard (e.g., 8): "))

checker = np.zeros((size, size), dtype=int)
checker[1::2, ::2] = 1
checker[::2, 1::2] = 1

print("\nCheckerboard pattern:")
print(checker)
```
