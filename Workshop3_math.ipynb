```python
import matplotlib.pyplot as plt

x = list(range(-10, 11))  # Values from -10 to 10
y = [i for i in x]  # y = x

plt.plot(x, y, marker='o', color='blue', label='y = x')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph of y = x')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 6, 8]
y = [3, 8, 1, 10]

plt.plot(x, y, marker='o', linestyle='-', color='red', label='Given points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Connecting Given Points')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 20)  # 20 values from -10 to 10

y1 = 5*x**3 + 2*x - 1
y2 = -2*x**3 + x**2 + 100
y3 = 2*np.pi*x + 20

plt.plot(x, y1, marker='o', label='5x^3 + 2x - 1', color='blue')
plt.plot(x, y2, marker='s', label='-2x^3 + x^2 + 100', color='red')
plt.plot(x, y3, marker='^', label='2x + 20', color='green')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Function Plots')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 20)

y1 = 5*x**3 + 2*x - 1
y2 = -2*x**3 + x**2 + 100
y3 = 2*np.pi*x + 20

plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(x, y1, marker='o', color='blue')
plt.title('y = 5x^3 + 2x - 1')
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(x, y2, marker='s', color='red')
plt.title('y = -2x^3 + x^2 + 100')
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(x, y3, marker='^', color='green')
plt.title('y = 2x + 20')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
vehicle_1 = [10, 20, 30, 40, 50]
vehicle_2 = [5, 15, 25, 35, 45]

plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.plot(days, vehicle_1, marker='o', color='blue', label='Vehicle 1')
plt.xlabel('Days')
plt.ylabel('Distance')
plt.title('Vehicle 1 Distance Covered')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(days, vehicle_2, marker='s', color='red', label='Vehicle 2')
plt.xlabel('Days')
plt.ylabel('Distance')
plt.title('Vehicle 2 Distance Covered')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
